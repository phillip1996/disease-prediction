```python
#### cleaning dataset to remove white space
import pandas as pd
cols=df_merge.columns
data=df_merge[cols].values.flatten()

s=pd.Series(data)
s.str.strip()
s=s.values.reshape(df_merge.shape)

df_merge=pd.DataFrame(s, columns=df_merge.columns)
df_merge.head()
```

```python
#### describing the data ###
df_merge.describe()
```

```python
### dependent and independent features
symptoms=df_merge[['Symptom_1','Symptom_2','Symptom_3','Symptom_4']]
y=df_merge['Disease']
```

```python
# ### changing text to integers
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

for i in symptoms.columns:
    symptoms[i]=le.fit_transform( symptoms[i].astype(str))
    

symptoms.head()
```

```python
#### import the machine learning library for splitting the model ###
from sklearn.model_selection import train_test_split
symptoms_train, symptoms_test, y_train, y_test = train_test_split(symptoms, y, test_size=0.2, random_state=101)
```

```python
from sklearn.metrics import accuracy_score
from sklearn import metrics
import pickle
from sklearn.linear_model import LogisticRegression
```

```python
###  training the data set
model=LogisticRegression()
model.fit(symptoms_train,y_train)
```

```python
###testing the model
prediction=model.predict(symptoms_test)
```

```python
accuracy_score(y_test,prediction)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,prediction)
```
